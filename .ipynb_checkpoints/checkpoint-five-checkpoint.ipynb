## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
# Importing naumpy and pandas libraries to read the data

# Supress Warnings
import warnings
warnings.filterwarnings('ignore')

# Import the numpy and pandas package
import numpy as np
import pandas as pd

# Read the given CSV file, and view some sample records
df = pd.read_csv("PLACES_County_Data_2021_Cleaned.csv")
df = df.dropna()
df
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
# Import matplotlib and seaborn libraries to visualize the data
import matplotlib.pyplot as plt 
import seaborn as sns
# Using pairplot we'll visualize the data for correlation
sns.pairplot(df, x_vars=['LPA_AdjPrev', 'BPMED_AdjPrev','BPHIGH_AdjPrev', 'HIGHCHOL_AdjPrev'], 
             y_vars='CHD_AdjPrev', size=4, aspect=1, kind='scatter')
plt.show()
```

```python
# Creating another df with fewer columns in order to see a heatmap that is readable
df2 = df[['LPA_AdjPrev', 'BPMED_AdjPrev','BPHIGH_AdjPrev', 'HIGHCHOL_AdjPrev', 'CHD_AdjPrev']]
df2
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
# I will be using CHD_AdjPrev and LPA_AdjPrev because I want to look at an outcome (CHD) and a risk factor (LPA)
# There are other columns with higher correlations, but they are both risk factors
x = df['LPA_AdjPrev']
y = df['CHD_AdjPrev']
```

```python
# Splitting the varaibles as training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
x_train
```

```python
y_train
```

```python
# Importing Statsmodels.api library from Stamodel package
import statsmodels.api as sm

# Adding a constant to get an intercept
x_train_sm = sm.add_constant(x_train)
```

```python
# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, x_train_sm).fit()

# Printing the parameters
lr.params
```

```python
# Performing a summary to list out all the different parameters of the regression line fitted
# The coefficient for LPA_AdjPrev is 0.1461 and a low p-value meaning it is significant
# F statistic p-value is low so the model fit is statistically significant
# 67.5% of the variance in CHD can be explained by LPA based on r squared
lr.summary()
```

```python
# Visualizing the regression line
plt.scatter(x_train, y_train)
plt.plot(x_train, 1.7882 + 0.1461*x_train, 'r')
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
# Predicting y_value using training data of s
y_train_pred = lr.predict(x_train_sm)

# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)
```

```python
# Plotting the histogram using the residual values
# the residuals are following a normal distribution with a mean 0
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
# Looking for any patterns in the residuals
plt.scatter(x_train,res)
plt.show()
```
